```python
import numpy as np
from numpy.linalg import inv
import matplotlib.pyplot as plt
#%matplotlib inline

#Test of stability
# step size
h = 0.1
# define numerical grid
t = np.arange(0,5.1,h)
# oscillation freq. of pendulum
w = 4

initialSolution = np.array([[1],[0]])

m_e = np.array([[1,h],[-w**2*h, 1]])

m_i = inv(np.array([[1, -h],[w**2*h, 1]]))

m_t = np.dot(inv(np.array([[1, -h/2], [w**2*h/2,1]])), np.array([[1,h/2],[-w**2*h/2, 1]]))

s_e = np.zeros((len(t), 2))
s_i = np.zeros((len(t), 2))
s_t = np.zeros((len(t), 2))

# do integrations
s_e[0, :] = initialSolution.T
s_i[0, :] = initialSolution.T
s_t[0, :] = initialSolution.T

for j in range(0, len(t)-1):
    s_e[j+1, :] = np.dot(m_e,s_e[j, :])
    s_i[j+1, :] = np.dot(m_i,s_i[j, :])
    s_t[j+1, :] = np.dot(m_t,s_t[j, :])

plt.figure(figsize = (12, 8))
plt.plot(t, s_e[:,0], 'b-')
plt.plot(t, s_i[:,0], 'g:')
plt.plot(t, s_t[:,0], 'r--')
plt.plot(t, np.cos(w*t), 'k')

plt.ylim([-3,3])
plt.xlabel('t')
plt.ylabel('$\Theta (t)$')
plt.legend(['Explicit', 'Implicit', 'Trapezoidal','Exact'])
plt.show()
```

```python
Week 5.1
Explicit shows unstable as motion increases with time
Implicit shows stable as motion decreases with time
Trapezoidal shows a neutral state as motion remains constant with time

Eulers explicit method is unstable with the step size of 0.1 as it diverges from the exact line. However when the step size is reduced to 0.001 or less then the method become significantly more stable and true to the exact line within the timeframe shown. 

Trapezoidal is the best method as it remains the closest to the exact over time.

The issue with the implicit solution is that the error propergates over time and causes damping of the system. 
```

```python
#ODE to be solved

def deriv(t,y):
    a = (3*t)/(1+t)
    B = 2*(1+t)**3*np.exp(-t)
    return (-a*y)+B

```

```python
#Runge-Kutta 4 
#IVP
y =  1.0 #initial value y (e.g. 1.0)
t =  0.1 #initial value x (e.g. 0.1)
dydt = deriv(t,y) 
steps =  round(14.9/h) #number of steps
h = 0.2 #step size

k1 = 0.0 #initialised
k2 = 0.0 #initialised
k3 = 0.0 #initialised
k4 = 0.0 #initialised
yout3 = np.zeros(steps+1)
tout3 = np.zeros(steps+1)
tout3[0] = t
yout3[0] = y

for i in range(1,steps+1): #Loop through the number of steps 
    k1 = h*dydt
    k2 = h*deriv(tout3[i-1]+h/2, yout3[i-1]+k1/2) #call function, note [i-1] gives previous value
    k3 = h*deriv(tout3[i-1]+h/2, yout3[i-1]+k2/2)
    k4 = h*deriv(tout3[i-1]+h, yout3[i-1]+k3)
    yout3[i] = yout3[i-1] + (k1/6) + (k2/3) + (k3/3) + (k4/6) #O(h**3) error 
    tout3[i] = tout3[i-1] + h # sets x value
    
    dydt = deriv(tout3[i],yout3[i]) #Evaluate ODE for the new position

plt.plot(tout3, yout3) #Line Plot of RK2 solver result
plt.title('RK4 Method')
plt.show()
```

```python
#Function for Jacobi rotation for each element in the p and q row
def Rotation(ai, ak, s, tau):
    gi=ai
    hk=ak
    ai = gi-s*(hk+gi*tau)
    ak = hk+s*(gi-hk*tau)

    return ai, ak
```

```python
#Cyclic Jacobi Method - as described in course handout
#Input A - numpy real symmetric matrix
def Jacobi(A=np.zeros(2,dtype=np.float64), maxsweeps=50):
    
    n=A.shape[0]
    d=np.zeros(n)   # array for storing eigenvalues
    v=np.identity(n) # array for storing eigenvectors

    z = np.zeros(n)
    b = np.copy(A.diagonal())
    d = np.copy(A.diagonal()) 

    nrot = 0 # number of rotations
    thresh = 0.0 # convergence threshold
    h=0.0
    t=0.0
    c=0.0
    s=0.0
    tau=0.0

    for i in range(1,maxsweeps+1):
        sm=0.0 # Sum of off diagonal elements.  REFINEMENT: In the first 3 sweeps this carrys out the rotation only if the determinant is greater than the threshold value. 
        for ip in range(n-1):
            for iq in range(ip+1, n):
                sm = sm + abs(A[ip,iq])
        
        if (sm == 0.0): return [d, v, nrot]
        if (i < 4):  
            thresh=0.2*sm/(n**2)
        else:
            thresh=0.0
        
        for ip in range(n-1):
            for iq in range(ip+1, n):
                g=100.0*abs(A[ip,iq])
                if (i > 4) and ((abs(d[ip])+g) == abs(d[ip])) and ((abs(d[iq]+g)) == abs(d[iq])):
                    #REFINEMENT: Skips jacobi rotation if the off diagonal element is small after four sweeps. 
                     A[ip,iq]=0.0                                                                 
                elif (abs(A[ip,iq]) > thresh):
                    h=d[iq]-d[ip]
                    if (abs(h)+g) == abs(h):
                        t=A[ip,iq]/h              
                    else:
                        theta=0.5*h/(A[ip,iq])
                        t=1.0/(abs(theta)+math.sqrt(1.0+theta*theta))
                        if theta < 0.0: t = -t
                    
                    c=1.0/math.sqrt(1+t**2)
                    s=t*c 
                    tau=s/(1.0+c)
                    h=t*A[ip,iq]
                    z[ip] = z[ip] - h
                    z[iq] = z[iq] + h
                    d[ip] = d[ip] - h
                    d[iq] = d[iq] + h
                    A[ip,iq] = 0.0

                    for j in range(ip):
                        A[j,ip],A[j,iq] = Rotation(A[j,ip],A[j,iq],s,tau)
                    for j in range(ip+1,iq):
                        A[ip,j],A[j,iq] = Rotation(A[ip,j],A[j,iq],s,tau)
                    for j in range(iq+1,n):
                        A[ip,j],A[iq,j] = Rotation(A[ip,j],A[iq,j],s,tau)
                    for j in range(n):
                        v[j,ip],v[j,iq] = Rotation(v[j,ip],v[j,iq],s,tau)
                    
                    nrot = nrot+1

        for ip in range(n):
            b[ip] = b[ip]+ z[ip]
            d[ip]=b[ip]
            z[ip]=0.0

    printf("Exceeded maximum allowed number of sweeps (%i)" %maxsweeps)  


```

```python
#Runge-Kutta 4 
#IVP
x =  0.0
y =  0.0 
z =  0.0 
dydx = f(x,y,z)
dzdx = g(x,y,z) 
steps =  10 #number of steps
h = 0.1 #step size

k1 = 0.0 #initialised
k2 = 0.0 #initialised
k3 = 0.0 #initialised
k4 = 0.0 #initialised
l1 = 0.0 #initialised
l2 = 0.0 #initialised
l3 = 0.0 #initialised
l4 = 0.0 #initialised
xout = np.zeros(steps+1)
yout = np.zeros(steps+1)
zout = np.zeros(steps+1)
xout[0] = x
yout[0] = y
zout[0] = z

for i in range(1,steps+1): #Loop through the number of steps 
    k1 = h*f(x,y,z)
    l1 = h*g(x,y,z)
    k2 = h*f(xout[i-1]+h/2, yout[i-1]+k1/2, zout[i-1]+l1/2) #call function, note [i-1] gives previous value
    l2 = h*g(xout[i-1]+h/2, yout[i-1]+k1/2, zout[i-1]+l1/2) #call function, note [i-1] gives previous value
    k3 = h*f(xout[i-1]+h/2, yout[i-1]+k2/2, zout[i-1]+l2/2) #call function, note [i-1] gives previous value
    l3 = h*g(xout[i-1]+h/2, yout[i-1]+k2/2, zout[i-1]+l2/2) #call function, note [i-1] gives previous value
    k4 = h*f(xout[i-1]+h, yout[i-1]+k3, zout[i-1]+l3) #call function, note [i-1] gives previous value
    l4 = h*g(xout[i-1]+h, yout[i-1]+k3, zout[i-1]+l3) #call function, note [i-1] gives previous value

    xout[i] = xout[i-1] + h 
    yout[i] = yout[i-1] + k1/6 +k2/3 + k3/3 +k4/6 
    zout[i] = zout[i-1] + l1/6 +l2/3 + l3/3 +l4/6 
    
    

plt.plot(xout,yout, label="z+x")
plt.plot(xout,zout, label="y**2 + x + z")
plt.legend()
plt.title('RK4 Method')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

import math 
%matplotlib inline
```

```python
#Add the three ODEs here as three separate, callable functions  

def dF1(F1, F2, F3, nout):
    return F2

def dF2(F1, F2, F3, nout):
    return F3

def dF3(F1, F2, F3, nout):
    return -F1*F3
```

```python
#Runge-Kutta for system of two first order ODEs 

def RK4(f3_0=0.5,dn=0.01,n=10): #initial values
    #x,y,z = initial values
    #h = step size
    #steps = number of steps
    steps=round(n/dn)
    h=dn

    K1 = 0.0
    K2 = 0.0
    K3 = 0.0
    K4 = 0.0

    l1 = 0.0
    l2 = 0.0
    l3 = 0.0
    l4 = 0.0

    m1 = 0.0
    m2 = 0.0
    m3 = 0.0
    m4 = 0.0 

    nout = np.zeros(steps+1)
    F1out = np.zeros(steps+1)
    F2out = np.zeros(steps+1)
    F3out = np.zeros(steps+1)

    #F1out[0] = f1
    #F2out[0] = f2
    F3out[0] = f3_0
    F3 = dF3(nout[0],F1out[0],F2out[0],F3out[0])
    F1 = dF1(nout[0],F1out[0],F2out[0],F3out[0])
    F2 = dF2(nout[0],F1out[0],F2out[0],F3out[0])
   
    for i in range(1,steps+1):
        K1 = h*F1
        l1 = h*F2
        m1 = h*F3
        K2 = h*dF1(nout[i-1]+h/2, F1out[i-1]+K1/2, F2out[i-1]+l1/2, F3out[i-1]+m1/2)
        l2 = h*dF2(nout[i-1]+h/2, F1out[i-1]+K1/2, F2out[i-1]+l1/2, F3out[i-1]+m1/2) 
        m2 = h*dF3(nout[i-1]+h/2, F1out[i-1]+K1/2, F2out[i-1]+l1/2, F3out[i-1]+m1/2)
        K3 = h*dF1(nout[i-1]+h/2, F1out[i-1]+K2/2, F2out[i-1]+l2/2, F3out[i-1]+m2/2)
        l3 = h*dF2(nout[i-1]+h/2, F1out[i-1]+K2/2, F2out[i-1]+l2/2, F3out[i-1]+m2/2)
        m3 = h*dF3(nout[i-1]+h/2, F1out[i-1]+K2/2, F2out[i-1]+l2/2, F3out[i-1]+m2/2)
        K4 = h*dF1(nout[i-1]+h,F1out[i-1]+K3,F2out[i-1]+l3,F3out[i-1]+m3)
        l4 = h*dF2(nout[i-1]+h,F1out[i-1]+K3,F2out[i-1]+l3,F3out[i-1]+m3)
        m4 = h*dF3(nout[i-1]+h,F1out[i-1]+K3,F2out[i-1]+l3,F3out[i-1]+m3)

        F3out[i] = F3out[i-1] + m1/6 + m2/3 + m3/3 + m4/6
        F2out[i] = F2out[i-1] + l1/6 + l2/3 + l3/3 + l4/6
        F1out[i] = F1out[i-1] + K1/6 + K2/3 + K3/3 + K4/6
        nout[i] = nout[i-1] + h

        
        F3 = dF3(nout[i],F1out[i],F2out[i],F3out[i])
        F1 = dF1(nout[i],F1out[i],F2out[i],F3out[i])
        F2 = dF2(nout[i],F1out[i],F2out[i],F3out[i])

      
    return F1out, F2out, F3out, nout
    
```

```python
#Run Runge-Kutta 4th order with wrapper of the shooting method for converging on 
#an acurate value for the intial value of f3(0)

f3initialprev = 1.0
f3initial = 0.5
F1o, F2o, F3o, n_out = RK4(1.0,0.01,10)
f2previous = F2o[-1]
F1o, F2o, F3o, n_out = RK4(0.5,0.01,10)
f2initial=F2o[-1]



#loop through '10' guess of f3(0) with step size as given in homework question 2, upto a value of 8 for the similarity variable (eta)
for i in range(8):
    f3new = f3initial + (((f3initial - f3initialprev)/(f2initial - f2previous))*(i-f2initial))
    F1o, F2o, F3o, n_out = RK4(f3new,0.01,6)
    print('iteration: %i, f3(0): %0.15f, f2(end): %0.15f' %(i, F3o[0], F2o[-1]))
    #check that the boundary condition at f2(eta) is equal to 1 (within a tolerance of 1e-18)
    if (abs(F2o[-1]-1) <= 1e-18):
        break

    #update f3(0)_(i), f3(0)_(i-1), f2(0)_(i), and f2(0)_(i-1) ready for next iteration
    
f3initialprev = f3initial
f3initial = f3new
f2previous = f2initial
f2initial = F2o[-1]


#Produce a single plot of f, f', and f'' against the similarity variable, eta

plt.plot(n_out, F1o, '--',label='f')
plt.plot(n_out, F2o, '-.', label='f\'')
plt.plot(n_out, F3o, label='f\'\'')
plt.legend()
plt.xlabel('eta')
plt.savefig('Blasius.pdf')
plt.plot(n_out, F3o)
plt.show()
```
